```python
def find_onering(origin_vertex):
    
    origin_halfedge = origin_vertex.halfedge
    
    onering_vertices = []
    halfedge_list = []
    
    cur_halfedge = origin_halfedge
    while True:
        halfedge_list.append(cur_halfedge)#helpful for debugging
        onering_vertices.append(cur_halfedge.opposite.vertex)
        cur_halfedge = cur_halfedge.opposite.next
        if cur_halfedge == origin_halfedge:
            break
            
    # print('Halfedge traversal: ',halfedge_list)
        
    return onering_vertices, halfedge_list


def get_dual(obj):
    # Create a new HalfedgeMesh for the dual
    dual_obj = HalfedgeMesh()

    # Get list of centroids of facets
    centroids = [get_centroids(facet) for facet in obj.facets] 
    dual_obj.update_vertices(centroids)

    # Create a facet for each vertex in the original mesh
    dual_obj.facets = [Facet(index=i) for i in range(len(obj.vertices))]

    # Create a halfedge for each vertex in the original mesh
    dual_obj.halfedges = [Halfedge(index=i) for i in range(len(obj.halfedges))]

    for vertex in obj.vertices:
        # Get one-ring neighbor vertices of the vertex (adjacent facets in the dual)
        one_ring, halfedge_list = find_onering(vertex)
        dual_obj.facets[vertex.index].update(halfedge=dual_obj.halfedges[vertex.halfedge.index])

        # Iterate over the one-ring neighbors
        for neighbor, near_he in zip(one_ring, halfedge_list):

            dual_obj.halfedges[near_he.index].update(vertex = dual_obj.vertices[near_he.facet.index],
                            next = dual_obj.halfedges[near_he.opposite.next.index],
                            facet = dual_obj.facets[vertex.index],
                            opposite = dual_obj.halfedges[near_he.opposite.index])
            
            current_half = dual_obj.halfedges[near_he.index]

            dual_obj.facets[neighbor.index].update(halfedge=current_half.opposite)

            dual_obj.vertices[current_half.vertex.index].update(halfedge=current_half)

    for vert in dual_obj.vertices:
        vert.x = -1*vert.x


    return dual_obj
```

```python
import igl
import numpy as np
from scipy.spatial import KDTree

## Load a mesh: v = [x,y,z] f = [v1.index, v2.index, v3.index]
v1, f1 = igl.read_triangle_mesh("bunny_v2/bun000_v2.ply")
v2, f2 = igl.read_triangle_mesh("bunny_v2/bun045_v2.ply")

n1 = igl.per_vertex_normals(v1, f1)
n2 = igl.per_vertex_normals(v2, f2)

def find_closest_points(kd_tree, points):
    distances, indices = kd_tree.query(points)
    return distances, indices

def compute_transformation(source_points, target_points):

    if len(source_points) != len(target_points):
        raise ValueError("Source and target point sets do not have the same number of points.")

    m = source_points.shape[1]

    centroid_source = np.mean(source_points, axis=0)
    centroid_target = np.mean(target_points, axis=0)

    # Center the points
    centered_source = source_points - centroid_source
    centered_target = target_points - centroid_target

    # Compute the covariance matrix
    H = np.dot(centered_source.T, centered_target)

    # Compute R
    U, S, Vt = np.linalg.svd(H)
    R = np.dot(Vt.T, U.T)

    # Ensure a right-handed coordinate system
    if np.linalg.det(R) < 0:
       Vt[-1, :] *= -1
       R = np.dot(Vt.T, U.T)
       
    t = centroid_target.T - np.dot(R, centroid_source.T)

    T = np.identity(m+1)
    T[:m, :m] = R
    T[:m, m] = t

    return T

def icp(source, normals_source, target, normals_target, max_iterations=1000, tolerance=1e-7, normal_threshold=np.cos(np.deg2rad(45))):
    source_points = source.copy()
    kd_tree = KDTree(target)
    prev_error = float('inf')
    

    for i in range(max_iterations):
        # Match each point pi in source to closest point qi in target
        _, indices = find_closest_points(kd_tree, source_points)
        target_points = target[indices]

        dot_products = np.einsum('ij,ij->i', normals_source, normals_target[indices])

        # Apply normal threshold
        valid = dot_products >= normal_threshold

        # Filter points
        valid_source_points = source_points[valid]
        valid_target_points = target[indices[valid]]

        if len(valid_source_points) == 0:
            raise ValueError("No valid points remaining after applying normal threshold.")

        # Compute R and t
        T = compute_transformation(valid_source_points, valid_target_points)

        # Apply transformation
        source_points_homogeneous = np.hstack((source_points, np.ones((source_points.shape[0], 1))))
        source_points= T.dot(source_points_homogeneous.T).T[:, :3] 

        # Error calculation and check for convergence
        squared_errors = np.sum((target_points - source_points)**2, axis=1)
        mean_error = np.mean(squared_errors)

        if np.abs(prev_error - mean_error) < tolerance:
            print(i)
            break
        prev_error = mean_error

    print(mean_error)
    print('calculate final transformation')

    return source_points

aligned_v2 = icp(v2, n2, v1, n1)
```

```python
def compute_weighted_transformation(source_points, target_points, weights):
    # Normalizing the weights
    normalized_weights = weights / np.sum(weights)

    # Compute the weighted centroid 
    centroid_source = np.dot(normalized_weights, source_points)
    centroid_target = np.dot(normalized_weights, target_points)

    centered_source = source_points - centroid_source
    centered_target = target_points - centroid_target

    # Compute the weighted covariance matrix
    W = np.diag(normalized_weights)
    H = centered_source.T @ W @ centered_target

    # Compute R
    U, S, Vt = np.linalg.svd(H)
    R = Vt.T @ U.T

    # Ensure a right-handed coordinate system
    if np.linalg.det(R) < 0:
        Vt[-1, :] *= -1
        R = Vt.T @ U.T

    t = centroid_target - R @ centroid_source

    # Construct the t
    m = source_points.shape[1]
    T = np.identity(m + 1)
    T[:m, :m] = R
    T[:m, m] = t

    return T


def icp_weight(source, target, max_iterations=1000, tolerance=1e-7):
    source_points = source.copy()
    kd_tree = KDTree(target)
    prev_error = float('inf')

    for i in range(max_iterations):
        # Match each point pi in source to closest point qi in target
        distances, indices = find_closest_points(kd_tree, source_points)
        target_points = target[indices]

        # Compute weights
        weights = 1 / (distances + 1e-6)  # Adding a small constant to avoid division by zero
        weights /= np.sum(weights)
        
        
        # Compute weighted transformation
        T = compute_weighted_transformation(source_points, target_points, weights)

        # Apply transformation
        source_points_homogeneous = np.hstack((source_points, np.ones((source_points.shape[0], 1))))
        source_points = T.dot(source_points_homogeneous.T).T[:, :3]

        # Error calculation and check for convergence
        squared_errors = np.sum((target_points - source_points)**2, axis=1)
        mean_error = np.mean(squared_errors)

        if np.abs(prev_error - mean_error) < tolerance:
            print(i)
            break
        prev_error = mean_error
        
    print(mean_error)

    return source_points

aligned_v2_weight= icp_weight(v2, v1)
```

```python
def rotation_matrix_z(angle_degrees):
    angle_radians = np.deg2rad(angle_degrees)
    cos_a = np.cos(angle_radians)
    sin_a = np.sin(angle_radians)
    return np.array([
        [cos_a, -sin_a, 0],
        [sin_a,  cos_a, 0],
        [0,      0,     1]
    ])

def error_calcualtion(v1, v2):
    squared_errors = np.sum((v1 - v2)**2, axis=1)
    mean_error = np.mean(squared_errors)

    return mean_error

def analyze_icp_convergence(v1, icp_function, angles):
    convergence_results = {}
    aligned_list = []
    rotated = []

    for angle in angles:
        # Create the rotated model
        R = rotation_matrix_z(angle)
        v2_rotated = np.dot(v1, R.T)

        n2 = igl.per_vertex_normals(v2_rotated, f2)
        
        # Run ICP 
        alligned_v1 = icp_function(v2_rotated, n2, v1, n1)
        aligned_list.append(alligned_v1)  
        rotated.append(v2_rotated)
        mean_error = error_calcualtion(alligned_v1, v1)
        
        convergence_results[angle] = {
            "mean_error": mean_error
        }

    return convergence_results, aligned_list, rotated

angles = [5, 10, 15, 20, 45, 90, 120, 180] 
results, aligned_list, rotated = analyze_icp_convergence(v1, icp, angles)
```

```python
def add_gaussian_noise(v, f, noise_scale_factor, target, target_normal):
    # Calculate the bounding box of M2
    min_vals = np.min(v, axis=0)
    max_vals = np.max(v, axis=0)
    bounding_box_dimensions = max_vals - min_vals

    # Determine the standard deviation
    noise_std = noise_scale_factor * bounding_box_dimensions

    # Generate Gaussian noise
    noise = np.random.normal(0, noise_std, v.shape)
    
    v_noisy = v + noise

    n = igl.per_vertex_normals(v_noisy, f)
    # Run ICP 
    alligned_v1 = icp(v_noisy, n, target, target_normal)
    
    return alligned_v1

noisy_v2_5 = add_gaussian_noise(rotated[0], f1, 1, v1, n1)
print(error_calcualtion(rotated[0], v1), error_calcualtion(noisy_v2_5, v1))
```

```python

def icp_subsample(source, source_normals, target, target_normals, max_iterations=1000, n_points=10000, tolerance=1e-7, normal_threshold=np.deg2rad(10), bins=40):
    source_points = source.copy()
    source_normals = source_normals.copy()
    prev_error = float('inf')

    for i in range(max_iterations):
        # 1: Normal-space sampling
        normal_bins = [[] for _ in range(bins)]
        # Calculate the angles in 2D normal space
        angles = np.arctan2(source_normals[:, 1], source_normals[:, 0])

        nan_indices = np.isnan(angles)
        inf_indices = np.isinf(angles)
        angles = angles[~(nan_indices | inf_indices)]

        bin_indices = np.floor((angles + np.pi) / (2 * np.pi) * bins).astype(int)


        bin_indices = np.clip(bin_indices, 0, bins-1)
        for idx, bin_idx in enumerate(bin_indices):
            normal_bins[bin_idx].append(idx)

        # Select a subset of points pi
        sampled_indices = []
        for bin in normal_bins:
            if len(bin) > 0:
                sampled_indices.extend(np.random.choice(bin, size=min(len(bin), n_points // bins), replace=False))
        sampled_source_points = source_points[sampled_indices]
        sampled_source_normals = source_normals[sampled_indices]

        # 2: Match each pi to closest point qi on other scan
        kd_tree = KDTree(target)
        _, indices = find_closest_points(kd_tree, sampled_source_points)
        target_points = target[indices]
        target_normals_matched = target_normals[indices]

        # 3: Reject “bad” pairs (pi, qi)
        normal_compatibility = np.einsum('ij,ij->i', sampled_source_normals, target_normals_matched) > np.cos(normal_threshold)
        valid_pairs = normal_compatibility

        filtered_source_points = sampled_source_points[valid_pairs]
        filtered_target_points = target_points[valid_pairs]

        # 4: Compute rotation R and translation t to minimize
        T = compute_transformation(filtered_source_points, filtered_target_points)  # Implement this function as needed

        # 5: Iterate after scan alignment: qi ← Rqi + t
        source_points_homogeneous = np.hstack((source_points, np.ones((source_points.shape[0], 1))))
        source_points = T.dot(source_points_homogeneous.T).T[:, :3]

        # Error calculation and check for convergence
        squared_errors = np.sum((target_points - sampled_source_points)**2, axis=1)
        mean_error = np.mean(squared_errors)

        if np.abs(prev_error - mean_error) < tolerance:
            break
        prev_error = mean_error

    return source_points

n1 = igl.per_vertex_normals(v1, f1)
n2 = igl.per_vertex_normals(v2, f2)

subsampled_aligned_v2 = icp_subsample(v2, n2, v1, n1)
```

```python
def combine_meshes(v1, f1, aligned_v2, f2):
    """
    Combines two meshes into a single mesh.

    Parameters:
    - v1: Vertices of the first mesh.
    - f1: Faces of the first mesh.
    - aligned_v2: Aligned vertices of the second mesh.
    - f2: Faces of the second mesh.

    Returns:
    - combined_v: Combined vertices.
    - combined_f: Combined faces.
    """
    # Combine the vertices
    combined_v = np.vstack((v1, aligned_v2))

    # The new faces for f2 need to offset the indices by the length of v1
    f2_offset = f2 + len(v1)

    # Combine the faces
    combined_f = np.vstack((f1, f2_offset))

    return combined_v, combined_f

v3, f3 = igl.read_triangle_mesh("bunny_v2/bun090_v2.ply")
n3 = igl.per_vertex_normals(v3, f3)

v4, f4 = igl.read_triangle_mesh("bunny_v2/bun180_v2.ply")
n4 = igl.per_vertex_normals(v4, f4)


v5, f5 = igl.read_triangle_mesh("bunny_v2/bun270_v2.ply")
n5 = igl.per_vertex_normals(v5, f5)

v6, f6 = igl.read_triangle_mesh("bunny_v2/bun315_v2.ply")
n6 = igl.per_vertex_normals(v6, f6)
```

```python
import scipy.linalg

def compute_transformation_point_to_plane(source_points, target_points, target_normals):
    if len(source_points) != len(target_points):
        raise ValueError("Source and target point sets do not have the same number of points.")
    m = source_points.shape[1]

    cross_products = np.cross(source_points, target_normals)
    concatenated_vectors = np.hstack((cross_products, target_normals))

    # Outer products for all pairs, and sum them to form matrix A
    A = np.einsum('ki,kj->ij', concatenated_vectors, concatenated_vectors)

    # Differences between target and source points
    point_differences = target_points - source_points

    # Dot products of differences and target normals
    dot_products = np.einsum('ki,ki->k', point_differences, target_normals)

    # Compute vector b
    b = np.einsum('k,ki->i', dot_products, concatenated_vectors)

    # Solve the linear system
    x = np.linalg.solve(A, b)

    # Extract rotation R and translation t from x
    R = scipy.linalg.expm(np.array([[0, -x[2], x[1]], [x[2], 0, -x[0]], [-x[1], x[0], 0]]))
    t = x[3:6]

    # Construct transformation matrix
    T = np.identity(m + 1)
    T[:m, :m] = R
    T[:m, m] = t

    return T


def icp_p2plane(source, source_normals, target, target_normals, max_iterations=1000, n_points=10000, tolerance=1e-7, normal_threshold=np.deg2rad(10), bins=40):
    source_points = source.copy()
    source_normals = source_normals.copy()
    prev_error = float('inf')

    for i in range(max_iterations):
        # 1: Normal-space sampling
        normal_bins = [[] for _ in range(bins)]
        angles = np.arctan2(source_normals[:, 1], source_normals[:, 0])

        # Check for NaN or Inf values in 'angles'
        nan_indices = np.isnan(angles)
        inf_indices = np.isinf(angles)
        angles = angles[~(nan_indices | inf_indices)]

        bin_indices = np.floor((angles + np.pi) / (2 * np.pi) * bins).astype(int)

        bin_indices = np.clip(bin_indices, 0, bins-1)
        for idx, bin_idx in enumerate(bin_indices):
            normal_bins[bin_idx].append(idx)

        # Select a subset of points pi
        sampled_indices = []
        for bin in normal_bins:
            if len(bin) > 0:
                sampled_indices.extend(np.random.choice(bin, size=min(len(bin), n_points // bins), replace=False))
        sampled_source_points = source_points[sampled_indices]
        sampled_source_normals = source_normals[sampled_indices]

        # 2: Match each pi to closest point qi on other scan
        kd_tree = KDTree(target)
        _, indices = find_closest_points(kd_tree, sampled_source_points)
        target_points = target[indices]
        target_normals_matched = target_normals[indices]

        # 3: Reject “bad” pairs (pi, qi)
        normal_compatibility = np.einsum('ij,ij->i', sampled_source_normals, target_normals_matched) > np.cos(normal_threshold)
        valid_pairs = normal_compatibility

        filtered_source_points = sampled_source_points[valid_pairs]
        filtered_target_points = target_points[valid_pairs]
        filtered_target_normal = target_normals_matched[valid_pairs]

        # 4: Compute rotation R and translation t to minimize
        T = compute_transformation_point_to_plane(filtered_source_points, filtered_target_points, filtered_target_normal)

        # 5: Iterate after scan alignment: qi ← Rqi + t
        source_points_homogeneous = np.hstack((source_points, np.ones((source_points.shape[0], 1))))
        source_points = T.dot(source_points_homogeneous.T).T[:, :3]

        # Error calculation and check for convergence
        squared_errors = np.sum(((target_points - sampled_source_points)* sampled_source_normals)**2, axis=1)
        mean_error = np.mean(squared_errors)

        if np.abs(prev_error - mean_error) < tolerance:
            print(i, mean_error)
            break
        prev_error = mean_error

    return source_points
```
